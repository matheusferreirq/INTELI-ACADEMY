5. Convertendo a coluna 'date_added' para datetime.


```python
df['date_added'] = pd.to_datetime(df['date_added'])

##verificando se foi convertida
df['date_added'].info()
```

6. Crie as colunas 'month_added' e 'year_added' a partir de 'date_added':


```python
df['month_added'] = df['date_added'].dt.month
df['year_added'] = df['date_added'].dt.year

## verificando se foram criadas
df.describe()
```

8. Filtrando somente os filmes e liste aqueles com duração maior que 100 minutos e que ao mesmo tempo foram lançados a partir de 2020:


```python
df['duration_num'] = df['duration'].str.extract(r'(\d+)').astype(int)

df_films100_2020 = df[(df['type'] == 'Movie') & (df['duration_num'] > 100) & (df['release_year'] >= 2020)]
df_films100_2020
```

9. Separando a coluna 'duration' em 'duration_int' (numérico) e 'duration_unit'
(unidade):

```python
df[['duration_int', 'duration_unit']] = df['duration'].str.extract(r'(\d+)\s*(\w+)')
df['duration_int'] = df['duration_int'].astype(int)

## verificando se a separação ocorreu
df
```

10. Criando a coluna ‘seasons_int’ e colocando o valor numérico da quantidade de
temporada contida em “duration” para as linhas TV Show.


```python
df['seasons_int'] = None
df.loc[df['type'] == 'TV Show', 'seasons_int'] = df.loc[df['type'] == 'TV Show', 'duration_num']
df['seasons_int']
```

11. Mostrando quais são a série com mais temporadas e o filme com maior tempo
de duração.


```python
more_seasons = df[df['type'] == 'TV Show'].sort_values(by='seasons_int', ascending=False).head(1)

more_duration = df[df['type'] == 'Movie'].sort_values(by='duration_int', ascending=False).head(1)
more_seasons
```

14.Formule uma hipótese subjetiva sobre o dataset.

```python
movies_df = df[df['type'] == 'Movie'].copy()
movies_df['short_movie'] = movies_df['duration_int'] < 90
short_movies_per_year = movies_df.groupby('year_added')['short_movie'].sum()

short_movies_per_year
```

R: Mediante análise do dataset, é possível analisar a crescente no número de filmes mais curtos, contendo duração de menos de 1h30min, ao longo dos anos, tangendo dos anos 2008 a 2021. Tal ocorrência pode ser justificada pelo aumento no uso da tecnologia e redes sociais, que podem refletir em uma possível tendência por contéudos mais rápidos de se consumir.
